```python
#red. 별오양. dash 선으로 sin 그리기
#green. 동그라미. 실선으로 cos 그리기
theta = np.linspace(0,2*np.pi,20)
print(theta)
plt.plot(theta,np.sin(theta),"r*--", label="sin")
#plt.figure() #새창 열기. 이거 없으면 겹쳐 그리기
plt.plot(theta,np.cos(theta),"go-", label="cos")
plt.xlabel(r"$₩theta$") #r은 raw
plt.ylabel("value")
plt.title("sin and cos", color="b")
plt.legend()
plt.grid()
plt.xlim([0,np.pi]) #0부터 pi까지 잘라줌.
plt.ylim([-1.5, 1.5])
plt.axis([0, np.pi, -1.5, 1.5])
```

```python
x=range(5)
y1=[i for i in range(1,6)]
y2=[i+100 for i in range(1,6)]
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)

plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
p1 = ax1.plot(x,y1, 'b')
p2 = ax2.plot(x,y2, 'r')
ax1.set_ylim([-10,10])
ax1.set_xlabel("x")
ax1.set_ylabel("y1", color="b")
ax2.set_ylabel("y2", color="r")
ax1.set_title("y1 and y2")
ax1.grid()
plt.legend(p1+p2,["y1","y2"], loc="right") #best는 알아서 위치를 좋은 곳에다가 놓아줌
```

## 2rㅐ 변수 함수, 등고선으로 보기, 3D로 보기

```python
a=np.linspace(-1,1,100)
b=np.linspace(-1,1,100)
A, B= np.meshgrid(a,b)
print(A)
print(B)

L = A**2 + B**2
plt.contour(a,b,L,20) #20 -> 등고선 개수. 5
plt.xlabel("a");plt.ylabel("b"); plt.grid()
```

```python
plt.figure(figsize=[10,9]) #[가로, 세로]
ax=plt.axes(projection="3d")
ax.plot_surface(A,B,L)

plt.figure(figsize=[10,9])
ax=plt.axes(projection="3d")
ax.plot_wireframe(A,B,L)
```

```python
figure=plt.figure(figsize=[10,9]) #[가로, 세로]
ax=plt.axes(projection="3d")
ax.view_init(elev=45,azim=-45)
ax.plot_surface(A,B,L)

figure=plt.figure(figsize=[10,9])
ax=plt.axes(projection="3d")
ax.view_init(elev=30,azim=-45)
ax.plot_wireframe(A,B,L)
```

```python
from matplotlib.colors import LightSource
plt.figure(figsize=[10,9])
ax = plt.axes(projection="3d")
ls=LightSource(altdeg=10,azdeg=-45)
rgb = ls.shade(L,plt.cm.RdYlBu)
ax.plot_surface(A,B,L, facecolors=rgb,
                linewidth=0, antialiased=False, shade=False)
```

```python
from matplotlib.colors import LightSource
plt.figure(figsize=[10,9])
ax = plt.axes(projection="3d")
ls=LightSource(altdeg=10,azdeg=-45)
rgb = ls.shade(L,plt.cm.gist_earth, vert_exag=0.1, blend_mode='soft')
ax.plot_surface(A,B,L, rstride=1, cstride=1, facecolors=rgb,
                linewidth=0, antialiased=False, shade=False)
```

```python
a=np.linspace(-1,1,10)
b=np.linspace(-1,1,10)
A, B= np.meshgrid(a,b)

L = A**2 + B**2

plt.figure(figsize=[10,9])
ax = plt.axes(projection="3d")
ax.plot(A.reshape(-1,1).squeeze(), B.reshape(-1,1).squeeze(), L.reshape(-1,1).squeeze(),"ro")

plt.figure(figsize=[10,9])
ax = plt.axes(projection="3d")
ax.scatter(A,B,L, s=100, c="g", marker="*",edgecolor="b", alpha=0.3)
```

```python
plt.figure(figsize=[10,9])
ax = plt.axes(projection="3d")
ax.plot_surface(A,B,L, alpha=0.3, cmap="viridis")
ax.scatter(A,B,L,s=100,c="r",marker="*")
```

##plotly

```python
import plotly.graph_objects as go

a=np.linspace(-2,2,100)
b=np.linspace(-2,2,100)
A,B = np.meshgrid(a,b)
L = np.cos(A**2.4)+np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(z=L, x=a, y=b, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title="cos(A^2.4)+sin(B^2.4)", width=700, height=600)
```

```python
fig = go.Figure(data=[go.Surface(z=L, x=a, y=b, opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title="cos(A^2.4)+sin(B^2.4)", width=700, height=600,
                  scene = dict(zaxis=dict(nticks=30, range=[-3,3])))
```

```python
im=test_DS.data[0]
plt.figure(figsize=(20,15))
plt.imshow(im, cmap="gray")
plt.colorbar()

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        plt.text(j,i, im[i,j].item(),
                 horizontalalignment="center",
                 color="k" if im[i,j] > torch.max(im)/1.5 else"w")

plt.xticks([])
plt.yticks([]) #plt.xticks=([]) 이렇게 하면 함수로 저장이 되면서 안 될 수 있음. 함수라는 걸 망각하면 안됨!
```
